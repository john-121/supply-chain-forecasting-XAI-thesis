- **Fix the Date column DTYPE**

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

print(df['Date'].dtypes)
```

- **Handling Missing Values**

```python
#Missing Values after the Date Feature fixed
print(df.isnull().sum().sort_values(ascending=False))
```

```python
# Drop rows where 'Date' is Not a time
df.dropna(subset=['Date'], inplace=True)
print(f"Remaining rows after dropping Not Dates: {len(df)}")
```

```python
# Impute numerical missing values using ffill and then bfill
for col in ['Units_Sold', 'Inventory_Level', 'Unit_Cost']:
    df[col] = df[col].fillna(method='ffill')
    df[col] = df[col].fillna(method='bfill')
```

```python
# Verify that all missing values have been handled
print("Missing Values Summary after handling:")
print(df.isnull().sum().sort_values(ascending=False))
```

```python
#Handle NaN values
print(df.isnull().sum().sort_values(ascending=False))
```

- **Uniqueness check Values in Categorical Columns**

```python
print("Unique Values in Categorical Columns")

categorical_cols = ['SKU_ID', 'Warehouse_ID', 'Supplier_ID', 'Region']

for col in categorical_cols:
    print(f"\nColumn: {col}")
    print(df[col].unique()[:20])  # show first 20 unique values
    print("Total unique:", df[col].nunique())
```

- **Standardize 'Region' Feature**

```python
# Standardize 'Region' column to fix typos
region_mapping = {
    'Noth': 'North',
    'Soth': 'South',
    'Est': 'East',
    'Wst': 'West'
}
df['Region'] = df['Region'].replace(region_mapping)

#Strip whitespace
categorical_cols = ['SKU_ID', 'Warehouse_ID', 'Supplier_ID', 'Region']

for col in categorical_cols:
    df[col] = df[col].astype(str).str.strip()

#Standardize case
df[col] = df[col].str.upper()

print("Unique regions after standardization:")
print(df['Region'].value_counts())
```

- **Handling for Outliers**

```python
# Calculate Profit
df['Profit'] = (df['Unit_Price'] - df['Unit_Cost']) * df['Units_Sold']

print("Statistical Summary ")
print(df.describe())

print("\n Negative Value Check ")
numeric_cols = ['Units_Sold', 'Inventory_Level', 'Supplier_Lead_Time_Days',
                'Reorder_Point', 'Order_Quantity', 'Unit_Cost', 'Unit_Price',
                'Demand_Forecast']

for col in numeric_cols:
    negatives = (df[col] < 0).sum()
    print(f"{col}: {negatives} negative values")


#Handling Outliers
print(" Capping Extreme Outliers ")

numeric_cols = ['Units_Sold', 'Inventory_Level', 'Supplier_Lead_Time_Days',
                'Reorder_Point', 'Order_Quantity', 'Unit_Cost', 'Unit_Price',
                'Demand_Forecast','Profit']

for col in numeric_cols:
    lower = df[col].quantile(0.01)
    upper = df[col].quantile(0.99)
    df[col] = df[col].clip(lower, upper)

```

Outliers were handled using selective quantile‑based capping (1st–99th percentile) on features with extreme spikes to improve the interpretability of EDA visualizations while preserving the natural distribution of the data

- **Handling Duplicate Rows**

```python
#Duplicate Rows Check
print("Number of duplicate rows:", df.duplicated().sum())

#Remove Duplicate Rows
df = df.drop_duplicates()
print("Duplicates removed.")
```

- **Time-Series Analysis: Sales and Profit Trends**

```python
# Aggregate daily sales and profit
daily_summary = df.groupby('Date').agg(
    total_units_sold=('Units_Sold', 'sum'),
    total_profit=('Profit', 'sum')
).reset_index()

plt.figure(figsize=(18, 6))

plt.subplot(1, 2, 1)
sns.lineplot(x='Date', y='total_units_sold', data=daily_summary)
plt.title('Total Units Sold Over Time')
plt.xlabel('Date')
plt.ylabel('Total Units Sold')

plt.subplot(1, 2, 2)
sns.lineplot(x='Date', y='total_profit', data=daily_summary)
plt.title('Total Profit Over Time')
plt.xlabel('Date')
plt.ylabel('Total Profit')

plt.tight_layout()
plt.show()
```

**Total Units Sold Over Time**: This graph displays the daily sum of units sold. We can observe
There are noticeable daily or weekly fluctuations, indicating periods of The consistent wave-like patterns suggest some level of **weekly seasonality**, where certain days of the week might consistently see higher or lower sales than others.

**Total Profit Over Time**: This graph shows the daily sum of profit.Like units sold, profit also experiences daily or weekly variations. This is expected, as profit is directly tied to the number of units sold.

**Correlation**: The pattern of the profit graph closely follows that of the units sold graph, reinforcing the strong positive correlation between units sold and profit that we might expect.

- **Seasonality Analysis of Units Sold**


First, I will group the DataFrame by 'Month' and calculate the average 'Units_Sold' for each month. Then, I will create a line plot to visualize these monthly averages, which will help identify any monthly seasonal patterns.



```python
df['Month'] = df['Date'].dt.month
monthly_avg_units_sold = df.groupby('Month')['Units_Sold'].mean()

plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_avg_units_sold.index, y=monthly_avg_units_sold.values, marker='o')
plt.title('Monthly Average Units Sold')
plt.xlabel('Month')
plt.ylabel('Average Units Sold')
plt.xticks(range(1, 13))
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Monthly Average Units Sold**: This graph illustrates the seasonal trends in sales over the course of a year. We can identify periods of higher and lower average sales volumes. For example, there might be distinct peaks during specific months (e.g., holiday seasons, product launch months) and troughs during other months. This helps in understanding long-term demand fluctuations and planning for annual cycles.



```python
df['DayOfWeek'] = df['Date'].dt.dayofweek # Monday=0, Sunday=6
daily_avg_units_sold = df.groupby('DayOfWeek')['Units_Sold'].mean()

plt.figure(figsize=(10, 6))
sns.lineplot(x=daily_avg_units_sold.index, y=daily_avg_units_sold.values, marker='o')
plt.title('Daily Average Units Sold')
plt.xlabel('Day of Week (0=Monday, 6=Sunday)')
plt.ylabel('Average Units Sold')
plt.xticks(range(7), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Daily Average Units Sold**: This graph reveals the weekly sales patterns, showing how average units sold vary from Monday to Sunday. As you noted, it highlights specific days that are typically busier or slower than others. For instance, the observation that Sunday is a peak sales day is a key takeaway. This granularity is essential for short-term operational planning, such as staffing, inventory replenishment, and daily marketing efforts.

```python
# Daily average profit
daily_profit = df.groupby('Day_Of_Week')['Profit'].mean()

plt.figure(figsize=(10,5))
plt.plot(daily_profit.index, daily_profit.values, marker='o')
plt.title('Daily Average Profit')
plt.xlabel('Day of Week (0=Monday, 6=Sunday)')
plt.ylabel('Average Profit')
plt.grid(True)
plt.show()

# Daily average demand forecast
daily_forecast = df.groupby('Day_Of_Week')['Demand_Forecast'].mean()

plt.figure(figsize=(10,5))
plt.plot(daily_forecast.index, daily_forecast.values, marker='o')
plt.title('Daily Average Demand Forecast')
plt.xlabel('Day of Week (0=Monday, 6=Sunday)')
plt.ylabel('Average Demand Forecast')
plt.grid(True)
plt.show()
```
